```python
assessments = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/assessments.csv')
courses = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/courses.csv')
student_assessments = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/studentAssessment.csv')
student_info = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/studentInfo.csv')
registration = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/studentRegistration.csv')
student_vle= pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/studentVle.csv')
vle = pd.read_csv('/kaggle/input/open-university-learning-analytics-dataset/vle.csv')

datasets = {
    'assessments':assessments,
    'courses':courses,
    'student_assessments':student_assessments,
    'student_info':student_info,
    'registration':registration,
    'student_vle':student_vle,
    'vle':vle}
```

```python
def basic_exploration(name, df):
    print("{}.head()".format(name))
    print(df.head())
    print("{}.shape".format(name))
    print(df.shape)
    print("\n{}.describe()".format(name))
    print(df.describe(include='all'))
    print("\n{}.isna().sum()".format(name))
    print(df.isna().sum())
    print("\n{}.value_counts()".format(name))
    print(df.value_counts())
    
    #matrix = df.corr()
    #plotting correlation matrix 
    #sns.heatmap(matrix, cmap="Greens", annot=True)
    print("\n")

#basic_exploration("assessments", assessments)
```

```python
for name, df in datasets.items():
    #(name)
    basic_exploration(name, df)
    
    
```
